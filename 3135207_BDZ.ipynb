# Doing something interesting with the reviews.csv file

# Using the data in the file we are going to perform a reviewer sentiment analysis that will enable the company to get an idea of how customers perceive the products. Which will help them make improvements, tailor marketing strategies, and understand what aspects customers like or dislike about their offerings. 

```python
# To achive this we need to read through the file first
```

```python
import pandas as pd
from textblob import TextBlob # which will help us carry out this analysis
```

```python
# we now load the file by reading through it
```

```python
# reading and loading the file
review_analysis = pd.read_csv('reviews.csv.csv')
```

```python
# Function to perform sentiment analysis on reviewer comments
def sentiment_analysis(text):
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    return sentiment_score

# Perform sentiment analysis on reviewer comments
review_analysis['Sentiment_Score'] = review_analysis['Review'].apply(sentiment_analysis)
review_analysis[['Uniq_id', 'Review', 'Sentiment_Score']]
```
